## 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
plt.xlabel('Цена')
plt.ylabel('Жилая площадь');
```

Стоимость дома прямо пропорциаональна его жилой площади

```python
sns.jointplot(x=df['price'], y=df['floors'], kind='reg');
plt.xlabel('Цена')
plt.ylabel('Количество этажей')
plt.title('Распределение цены от этажности');
```

Наиболее дорогими в большинстве случаев являются двухэтажные дома, но самый дорогой дом имеет 2.5 этажа

```python
sns.jointplot(x=df['condition'], y=df['floors'], kind='reg');
plt.xlabel('Состояние')
plt.ylabel('Количество этажей')
plt.title('Зависимость состояния от количества этажей');
```

Все здания, имеющие более 2-х этажей, не имеют плохой оценки состояния (удовлетворительно и выше)

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'));
plt.xlabel('Цена')
plt.ylabel('Вид из окна')
plt.title('Зависимость цены от вида из окна');
```

Все здания, имеющие более 2-х этажей, не имеют плохой оценки состояния (удовлетворительно и выше)

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Оценка дизайна')
plt.title('Зависимость цены от дизайна');
```
